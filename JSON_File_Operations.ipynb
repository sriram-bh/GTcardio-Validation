```python
# look at whatever noise frequency you want. find the frequency 
# number (freq_spectum, or fftarr index that produces that peak). 
# set the value at that index to 0, thereby eliminating the peak. 
# Then, do an inverse fft to reproduce original data
```

```python
x = np.arange(0,12*np.pi,.1)   # start,stop,step
ysin = np.sin(x)
ycos = np.cos(0.3*x)
y = []
for i in range(len(ysin)):
    y.append(ysin[i] + ycos[i])
plt.plot(x, y)
```

```python
freq_arr = np.fft.fft(y)
freq_sin = np.fft.fft(ysin)
length = len(freq_arr)
precision = 6

for i in range (precision):
    freq_arr[i] = 0
    freq_arr[length-1-i] = 0


# plt.plot(freq_arr[0:precision])
# plt.plot(freq_sin[0:precision])

plt.plot(y)
plt.plot(np.fft.ifft(freq_arr))
```
